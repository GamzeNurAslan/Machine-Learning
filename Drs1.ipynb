```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 40, 1000)

d_r = 2.5 * t
d_s = 3 * (t-5)

fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)        
y1 = 3 * x                              
y2 = 1 + (5 * x) / 2                    

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

ax.set_xlim([0, 3])
ax.set_ylim([0, 8])

ax.plot(x, y1, c='green', label='y = 3x')
ax.plot(x, y2, c='brown', label='y = 1 + (5x)/2')

plt.axvline(x=2, color='purple', linestyle='--')
plt.axhline(y=6, color='purple', linestyle='--')

plt.legend()
plt.grid(True)
plt.title('Dorularn Kesiimi')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)  # x tanmland

y1 = -5 + (2 * x) / 3
y2 = (7 - 2 * x) / 5

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# X ve Y eksenlerini gri izgilerle ekleyelim
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

# Grnm aral
ax.set_xlim([-2, 10])
ax.set_ylim([-6, 4])

# Dorular iziliyor
ax.plot(x, y1, c='green', label='y1 = -5 + (2x)/3')
ax.plot(x, y2, c='brown', label='y2 = (7 - 2x)/5')

# Kesiim noktas: x = 6, y = -1
plt.axvline(x=6, color='purple', linestyle='--')
plt.axhline(y=-1, color='purple', linestyle='--')

# Aklama ve zgara
plt.legend()
plt.grid(True)
plt.title('ki Dorunun Kesiimi')
plt.show()
```

* tf.Variable : Deeri deitirilebilir tensr.
* tf.constant : Deeri sabit, deitirilemeyen tensr.
* tf.placeholder : TF 1.xte vard. Eitim srasnda dardan veri almak iin kullanlrd. TF 2.xte yerine tf.function ve tf.data kullanlabilir.
* tf.SparseTensor : Seyrek (ou deeri 0 olan) verileri verimli temsil eder.

```python
import tensorflow as tf

x_tf = tf.Variable(25, dtype=tf.int16)  
y_tf = tf.Variable(3, dtype=tf.int16)    

tf_sum = tf.add(x_tf, y_tf)

np_result = tf_sum.numpy()

tf_float = tf.Variable(25., dtype=tf.float16)

print("x_tf:", x_tf)
print("y_tf:", y_tf)
print("Toplam (Tensor):", tf_sum)
print("Toplam (NumPy):", np_result)
print("NumPy tipi:", type(np_result))
print("Ondalkl tensr:", tf_float)
```

```python
import numpy as np
import matplotlib.pyplot as plt

b = np.linspace(-10, 10, 1000)  
c1 = 2 - 2 * b
c2 = (7 - 5 * b) / 3

fig, ax = plt.subplots()
plt.xlabel('b', c='darkorange')
plt.ylabel('c', c='brown')

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 3])
ax.set_ylim([-1, 5])

ax.plot(b, c1, c='purple', label='c1 = 2 - 2b')
ax.plot(b, c2, c='orange', label='c2 = (7 - 5b)/3')

plt.axvline(x=-1, color='green', linestyle='--')
plt.axhline(y=4, color='green', linestyle='--')

plt.legend()
plt.title("Dorularn Kesiimi")
plt.grid(True)
plt.show()
```
